```python
# github.com/echtr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import random
```

```python
random_price = lambda price_ : price_*4 + random.randint(1, 1000)
area = [i*10 for i in range(50)]
price = [random_price(i) for i in area]


df = pd.DataFrame({"area": area,
                 "price": price}).iloc[1:, :]
#pd.DataFrame(df)
```

```python
plt.scatter(df.area, df.price)
plt.show()
```

```python
x = df["area"].values.reshape(-1, 1)
y = df["price"].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3)
model = LinearRegression()
model.fit(x_train, y_train)
prediction = model.predict(x_test)
print(prediction)
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.scatter(x_train, y_train, color = "red") # bunu kullanabilirsin - train model
plt.scatter(x_test, y_test, color = "green") # bunu da kullanabilirsin - test model
plt.plot(x_train, model.predict(x_train), color = "blue") #burası hep sabit
plt.xlabel("area")
plt.ylabel("price")
plt.legend(["Linear Regression Model", "Train Model", "Test Model"])
plt.show()
```
